### Matplotlib example workflow

```python
# 0. import matplotlib and get it redy for plotting in jupyter
import matplotlib.pyplot as plt
%matplotlib inline

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #width and heigth 

# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(title= 'Simple Plot',
       xlabel= 'x-axis',
       ylabel= 'y-labal')

# 5. save and show (you save the whole figure)
fig.savefig("images\sample-plot.png")

```

```python
fig, ax = plt.subplots(figsize=(5, 5))

data = np.random.randint(10, size=(2, 8))
x = data[0]
y = data[1]

ax.plot(x, y)

ax.set(title= 'random graph',
       xlabel= 'Random x',
       ylabel= 'Random y')

plt.show()
```

```python
# making a plot from dictionary
nut_butter_prices= {'Almond butter': 10,
                    'Peanut butter': 8,
                    'Cashew butter': 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title= "Dan's Nut Butter Store",
       ylabel= "Prices($)")
```

```python
# make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two Options for subplots

```python
# subplot Option 1
fig, ((ax1, ax2), (ax3, ax4))= plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize= (10, 5))

#plot to different axis
ax1.plot(x, x/2);

y = np.linspace(0, 360, 45)
ax2.scatter(y, np.cos(y));

ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());

ax4.hist(np.random.randn(1000));
```

```python
# subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize= (10, 5))

#plot to each different axis
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs Matplotlib OO method)

* when plotting something quickly, okay to use the pyplot

* when plotting something more advanced, use the OO method

```python
heart_disease.head(10)
```

```python
over_50 = heart_disease[heart_disease.age > 50]
over_50.head()
```

```python
over_50.info()
```

```python
# pyplot method
over_50.plot(kind= 'scatter',
             x= 'age',
             y= 'chol',
             c= 'target',);
```

```python
## OO Method mixed with pyplot

fig, ax = plt.subplots(figsize= (10, 6))
over_50.plot(kind= 'scatter',
             x= 'age',
             y= 'chol',
             c= 'target',
             ax=ax);

#ax.set_xlim([45, 100])
```

```python
## OO Method from scratch
fig, ax = plt.subplots(figsize= (10, 6))

# plot the data
scatter = ax.scatter(x= over_50.age,
                     y= over_50.chol,
                     c= over_50.target);

# customizing the plot
ax.set(title= "Heart diseases and Cholesterol Levels",
       xlabel= "Age",
       ylabel= 'Cholesterol');

# add a legend
ax.legend(*scatter.legend_elements(), title= "Target")
```
