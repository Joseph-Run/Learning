### Matplotlib example workflow

```python
# 0. import matplotlib and get it redy for plotting in jupyter
import matplotlib.pyplot as plt
%matplotlib inline

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #width and heigth 

# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(title= 'Simple Plot',
       xlabel= 'x-axis',
       ylabel= 'y-labal')

# 5. save and show (you save the whole figure)
fig.savefig("images\sample-plot.png")

```

```python
fig, ax = plt.subplots(figsize=(5, 5))

data = np.random.randint(10, size=(2, 8))
x = data[0]
y = data[1]

ax.plot(x, y)

ax.set(title= 'random graph',
       xlabel= 'Random x',
       ylabel= 'Random y')

plt.show()
```
